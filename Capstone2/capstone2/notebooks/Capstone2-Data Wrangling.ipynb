```python
#code found online to display directory. Directory created via cookiecutter package within a conda virtual environment 'capstone2'
def list_files(startpath):
    for root, dirs, files in os.walk(startpath):
        level = root.replace(startpath, '').count(os.sep)
        indent = ' ' * 4 * (level)
        print('{}{}/'.format(indent, os.path.basename(root)))
        subindent = ' ' * 4 * (level + 1)
        for f in files:
            print('{}{}'.format(subindent, f))
```

# Data Cleaning

```python
#count NaN Values in each column with NaN
cols = df[emp].isnull().sum()
cols
```

```python
#filter for list of columsn with less than 2000 missing values ~0.1 % entries
cols =cols [cols<2000]
ls = cols.index.to_list()
ls
```

```python
#drop all row with Null values in columns with less than 2000 total nulls (filtered by above)
df = df.dropna(subset=['1st_Road_Number',
 'Did_Police_Officer_Attend_Scene_of_Accident',
 'Latitude',
 'Location_Easting_OSGR',
 'Location_Northing_OSGR',
 'Longitude',
 'Pedestrian_Crossing-Human_Control',
 'Pedestrian_Crossing-Physical_Facilities',
 'Speed_limit',
 'Time',
 'InScotland',
 'Vehicle_Location.Restricted_Lane'])
```

```python
df.info(show_counts=True)
```

```python
#create list of columns with null values()

emp =[]
for col in  df.columns:
    if (df[col].isnull().sum()) >0:
        emp.append(col)
df[emp].info(show_counts=True)
```

```python
#impute most common value
df['Age_of_Vehicle'] = df['Age_of_Vehicle'].fillna(df['Age_of_Vehicle'].mode()[0])
df['Driver_IMD_Decile'] = df['Driver_IMD_Decile'].fillna(df['Driver_IMD_Decile'].mode()[0])
df['Propulsion_Code'] = df['Propulsion_Code'].fillna(df['Propulsion_Code'].mode()[0])
df['Engine_Capacity_.CC.'] = df['Engine_Capacity_.CC.'].fillna(df['Engine_Capacity_.CC.'].mode()[0])
df['2nd_Road_Class'] = df['2nd_Road_Class'].fillna(df['2nd_Road_Class'].mode()[0])
df['model'] = df['model'].fillna(df['model'].mode()[0])
df['2nd_Road_Number'] = df['2nd_Road_Number'].fillna(df['2nd_Road_Number'].mode()[0])
```

```python
df[emp].info(show_counts=True)
```

```python
#impute missing Make as "Missing"
df['make'].fillna("MISSING", inplace=True)
```

```python
#drop entire columns - this is duplicate of lat and lon
df.drop(columns = 'LSOA_of_Accident_Location', inplace=True)
df.drop(columns = ['Location_Easting_OSGR',
 'Location_Northing_OSGR'], inplace=True)
```

```python

kv ={}
#ls1 = df['Weather_Conditions'].value_counts().head(10).index.to_list()
#kv['Weather_Conditions']= ls1

for i in df.columns:
    kv[i] = df[i].value_counts().head(10).index.to_list()
```
